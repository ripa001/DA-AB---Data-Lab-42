##### Ex. 6

```python
img = mpimg.imread("img/IMGP9931-5.jpg")
img2 = mpimg.imread("img/IMGP0228-17.jpg")
```

```python
plt.imshow(img)
```

```python
def mulimg(img, matrix):
    img2 = img.reshape((img.shape[0] * img.shape[1], 3))
    # output = np.matmul(img2, matrix)
    output = np.dot(img2, matrix)
    return output.reshape(img.shape).astype(img.dtype)
```

```python
mat = np.concatenate((arr_n(3), arr_n(3), arr_n(3)))
linear = mulimg(img, mat)
```

```python
plt.imshow(linear)
```

```python
x1 = np.arange(9.0).reshape((3, 3))
```

```python
identity = mulimg(img, x1)   
```

```python
plt.imshow(identity)
```

```python
x1 = np.arange(9).reshape((3, 3))
x2 = np.arange(3.0)
x3 = x1 * x2
```

```python
boh = mulimg(img, x3)
```

```python
plt.imshow(boh)
```

```python
x4 = np.random.randn(3,3)
```

```python
boh = mulimg(img, x4)
```
